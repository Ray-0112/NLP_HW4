```python
import os

list_folders = ['AA', 'AB', 'AC', 'AD', 'AE', 'AF', 'AG', 'AH', 'AI', 'AJ', 'AK', 'AL', 'AM']
mypath = "D:/wiki_zh"
infos = []
for folder in list_folders:
    foldername = os.path.join( mypath, folder)
    list_files = os.listdir( foldername)
    temp = []
    info = {}
    for file in list_files:
        # 產生檔案的絕對路徑
        info['目錄'] = folder
        fullpath = mypath + "/" +folder + "/" + file
        #print(fullpath)
        if os.path.isfile( fullpath):
            temp.append(fullpath)
    info['檔案'] = temp
    infos.append(info)
print(infos)
```

```python
stopwords_path='C:/Users/Ray/Desktop/NLP_HW4/stopwords.txt'

stopwords = []
with open(stopwords_path, 'r', encoding='utf-8') as f:
    for line in f:
        if len(line)>0:
            stopwords.append(line.strip())
```

```python
import json
import re
from os import walk
from os import listdir
from os.path import isfile, isdir, join


def extract(path):
    lis = []
    text = ""
    with open(path ,encoding = "utf-8") as f:
        for i,line in enumerate(f.readlines()):
            j = json.loads(line)
            #print(j["text"])
            temp = j["text"]
            temp.strip('\n')
            temp = re.sub("[A-Za-z0-9\：\〈\〉\<\>\-\『\』\？\；\·\—\，\「\」\、\（\）\，\《\》\,\。\“ \”\.]", "", temp)
            lis.append(temp)
            text = text + temp
    f.close()
    return text

f = open(path, 'w',encoding = 'utf-8')
```

```python
output_path = 'all_cut_word.txt'
input_path = 'all_content.txt'

def tokenizer(s):
    words = []
    cut = jieba.cut(s)
    for word in cut:
        if word not in stopwords:
            words.append(word)
    return words

with open(output_path,'w',encoding='utf-8') as o:
    with open(input_path, 'r', encoding='utf-8') as f:
        for line in f:
            s=tokenizer(line.strip())
            o.write(" ".join(s)+"\n")
print("done!!!")
```

```python
input_path = 'all_cut_word.txt'
output_path = 'all_word_cut.txt'

with open(output_path,'w',encoding='utf-8') as o:
    with open(input_path, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            o.write(' '+line)
o.close()
f.close()
```

```python
from gensim.models import word2vec, fasttext
import logging

def train(data):
    # Settings
    #logging.basicConfig(level=logging.INFO)
    seed = 666
    sg = 0
    window_size = 10
    vector_size = 100
    min_count = 1
    workers = 8
    epochs = 5
    batch_words = 10000

    train_data = word2vec.LineSentence(data)
    model = word2vec.Word2Vec(
        train_data,
        min_count=min_count,
        vector_size=vector_size,
        workers=workers,
        epochs=epochs,
        window=window_size,
        sg=sg,
        seed=seed,
        batch_words=batch_words
    )

    model.save('word2vec.model')
    print("done!!!")
    
train('all_word_cut.txt')
```
